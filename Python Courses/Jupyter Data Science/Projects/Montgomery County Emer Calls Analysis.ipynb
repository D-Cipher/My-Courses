## Details for Call Reasons

In the titles column there are "Reasons/Departments" specified before the title code. These are EMS, Fire, and Traffic. For example, if the title column value is EMS: BACK PAINS/INJURY, the Reason column value would be EMS. These string values are going to be extracted with the following code into two columns: "Reason" and "Department".

```python
titleSplit = df["title"].str.split(": ", expand=True)
titleSplit.columns = ["department", 'reason']

df2 = df.join(titleSplit)

df2.head()
```

** Most common Reason for a 911 call: **

```python
df2["department"].value_counts().nlargest(5)
```

```python
sns.countplot(x="department", data=df2, palette="viridis")
```

### Call Day Time Analysis

Now we will use Jupyter's tab method to explore the various attributes you can call. We will make timestamp into DateTime objects, use .apply() to create 3 new columns called Hour, Month, and Day of Week. You will create these columns based off of the timeStamp column, reference the solutions if you get stuck on this step.

```python
#===Convert timeStamp to pd datetime

df2["timeStamp"] = pd.to_datetime(df2["timeStamp"])

type(df2["timeStamp"].iloc[0])
```

```python
df2["hour"] = df2["timeStamp"].apply(lambda time: time.hour)

df2["month"] = df2["timeStamp"].apply(lambda time: time.month)

df2["week"] = df2["timeStamp"].apply(lambda time: time.dayofweek)

dmap = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}

df2["week"] = df2["week"].map(dmap)

df2
```

** Calls Breakdown by Week **

```python
sns.countplot(x="week", data=df2, hue="department", palette="viridis")

# To relocate the legend
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

** Calls Breakdown by Month **

```python
sns.countplot(x="month", data=df2, hue="department", palette="viridis")

# To relocate the legend
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

Since there are some missing Months, we can maybe this information by plotting the information by a simple line plot that fills in the missing months, in order to do this, we'll need to create a gropuby object called byMonth, where we can group the DataFrame by the month column and use the count() method for aggregation.

```python
byMonth = df2.groupby("month").count()
byMonth
```
